## Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt

def f(p):
    n = 100
    y = np.zeros(n)
    y[int(n * p):] = 1
    return y
    
N = 1000
n = 30

F = np.random.choice(f(.5), (N,n))
nf = np.sum(F, axis=1)

fig = plt.figure()

plt.subplot(2,1,1)

plt.title('f')
plt.plot(f(.5))

plt.subplot(2,1,2)
plt.hist(nf)
plt.title('n*f')
plt.xlabel(f"<n*f> = {np.mean(nf)}")
fig.tight_layout()
plt.show()
```

```python
n_vars = 30
f_list = np.zeros((n_vars, 100))
F = np.zeros((N,n_vars))
for i in range(n_vars):
    f_list[i,:] = f(np.random.rand())
    F[:,i] = np.random.choice(f_list[i,:], N)

    
nf = np.sum(F, axis=1)
fig = plt.figure()

plt.subplot(2,1,1)
plt.title('F')
plt.plot(np.transpose(f_list))

plt.subplot(2,1,2)
plt.hist(nf)
plt.title('F')
plt.xlabel(f"<F> = {np.mean(nf)}")
fig.tight_layout()
plt.show()
```
