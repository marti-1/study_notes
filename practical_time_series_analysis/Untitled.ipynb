```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

In a (weak) stationary time series, there is no:
* systematic change in mean (no trend)
* systematic change in variance
* periodic variations

**Autocovariance**

$$
c_k = \frac{\sum^{N-k}_{t=1}(x_t - \bar{x})(x_{t+k}-\bar{x})}{N}
$$

**Autocorrelation**

$$
r_k = \frac{c_k}{c_0} \\
r_k = \frac{\sum^{N-k}_{t=1}(x_t - \bar{x})(x_{t+k}-\bar{x})}{\sum^N_{t=1}(x_t - \bar{x})^2}
$$

## Random walk

### Brownian motion

```python
random_process = np.cumsum(np.random.normal(0,1,100))
plt.plot(random_process)
```

```python
sm.graphics.tsa.plot_acf(random_process)
plt.show()
```

the above is clearly non-stationary process. We can get a stationary process in this case by subtracting the previous value from current value:

$$
X_t - X_{t-1} = N(0,1)
$$

```python
plt.plot(np.diff(random_process))
```

```python
sm.graphics.tsa.plot_acf(np.diff(random_process))
plt.show()
```
