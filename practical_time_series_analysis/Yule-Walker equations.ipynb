# Yule-Walker equations

The goal is to find $p(k)$
$$
p(k)=\frac{\gamma(k)}{\gamma(0)} \\
\gamma(x)=\gamma(-k)
$$

We have AR(2) process:

$$
x_{t}=\frac{1}{3} x_{t-1}+\frac{1}{2} x_{t-2}+z_{t}
$$

Is it stationary? (Why do we care? Because by the definition of stationary process $\gamma(t,t+k)=\gamma(k)$ -- only distance that matters and not a specific location).

$$
x_{t}-\frac{1}{3} X_{t-1}+\frac{1}{2} X_{t-2}=Z_{t} \\
x_{t}\left(1-\frac{1}{3} B +\frac{1}{2} B^{2}\right)=Z_{t} \\
1-\frac{1}{3} B+\frac{1}{2} B^{2}=0 \\
B=\frac{-2 \pm \sqrt{76}}{6} \\
|B|>1
$$
Therefore, stationary.
$$
E\left[x_{t}\right]=\mu \\
E\left[X_{t}\right]=E\left[\frac{1}{3} X_{t-1}\right]+E\left[\frac{1}{2} X_{t-2}\right]+E\left[Z_{t}\right] \\
\mu=\frac{1}{3} \mu+\frac{1}{2} \mu \\
\mu=0
$$

$\gamma(k) = E[X_t,X_{t+k}] = E[X_t,X_{t-k}]$

Multiply both sides with $X_{t-k}$:

$$
\gamma(-k)=\frac{1}{3} \gamma(-k+1)+\frac{1}{2} \gamma(-k+2)+0
$$

Where does $-k + 1$ come from?

$$
\begin{array}{l}
k=3 \\
\gamma(t-3, t-1) \\
t=10 \\
(7,9) \rightarrow \gamma(-k+1) \\
\end{array}
$$

$$
\begin{aligned}
\gamma(k)=\frac{1}{3} \gamma(k-1)+\frac{1}{2} \gamma(k-2) && / \gamma(0) \\
p(k)=\frac{1}{3} p(k-1)+\frac{1}{2} p(k-2)
\end{aligned}
$$

Solve the difference equation:
$$
\lambda^{n}=\frac{1}{3} \lambda^{n-1}+\frac{1}{2} \lambda^{n-2} \\
\lambda^{n}-\frac{1}{3} \lambda^{n-1}-\frac{1}{2} x^{n-2}=0 \\
x^{2}-\frac{1}{3} x-\frac{1}{2}=0 \\
\lambda_{1}=\frac{2+\sqrt{76}}{12}, \quad \lambda_{2}=\frac{2-\sqrt{76}}{12} \\
p(k)=c_{1}\left(\frac{2+\sqrt{76}}{12}\right)^{k}+c_{2}\left(\frac{2-\sqrt{76}}{12}\right)^{k}
$$

Use constraints to obtain coefficients:

$$
\begin{aligned}
p(0) &=1 \Rightarrow c_{1}+c_{2} =1 \\
p(1) &=\frac{1}{3} p(0)+\frac{1}{2} p(-1), && p(1)=p(-1) \\
p(1) &=\frac{1 / 3}{1-1 / 2}=\frac{2}{3} \\
\left\{\begin{array}{l}c_{1}+c_{2}=1 \\ c_{1}\left(\frac{2+\sqrt{76}}{12}\right)+c_{2}\left(\frac{2-\sqrt{76}}{12}\right)=\frac{2}{5}\end{array}\right. \\
c_{1}=\frac{4+\sqrt{6}}{8}, \quad c_{2}=\frac{4-\sqrt{6}}{8}
\end{aligned}
$$

Finally:

$$
p(k)=\frac{4+\sqrt{6}}{8}\left(\frac{2+\sqrt{76}}{12}\right)^{k}+\frac{4-\sqrt{6}}{8}\left(\frac{2-\sqrt{76}}{12}\right)^{k}
$$
